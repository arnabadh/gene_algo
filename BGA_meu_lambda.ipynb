# Main function of genetic algorithm ::

 ## Variables:  Dimension:   Used for:
    no_of_var: 1                number of independent variables
    N:         1                number of population
    x_max:     no_of_var        max value of x1,x2,x3...
    x_min:     no_of_var        min value of x1,x2,x3...
    eps:       1                epsilon, used for accuracy
    length:    calculated       no. of bits representing x1,x2,x3...
    cumm_len:  same as length   ----
    S:         Nxsum(length[])  solution space
    decval_x:  Nxno_of_var      decoded value corresponding to x1,x2,x3...
    x:         Nxno_of_var      value of x1,x2,x3...corresponding to each population

```python
#change the definition here
fitness_eval = lambda param: (-1*param) 
```

```python
f1=open('input.txt','r')
inlist=f1.readlines()
gen=0

#Enter maximum generation:
maxgen = 200
no_of_var=int(inlist[0])
#Enter number of population:
N=60

x_max=[]
x_min=[]
length=[]
cumm_len=[]
cumm_len.append(0)
i=0

j=1
while i < no_of_var:
    temp1=float(inlist[j])
    temp2=float(inlist[j+1])

    if temp2>=temp1 :
        print("Incorrect range of x!")
        continue
    x_max.append(temp1)
    x_min.append(temp2)

    eps=float(inlist[j+2])
    temp=math.ceil(math.log((x_max[-1]-x_min[-1])/eps)/math.log(2))
    length.append(temp)
    cumm_len.append(temp+cumm_len[-1])
    i=i+1
    j=j+3

tot_bit=sum(length)
S=np.random.randint(0, 2, size=(N,tot_bit ))
S_new=np.zeros((N,tot_bit))


    
while gen<maxgen:
    gen=gen+1
    
    
    
    Par_off=np.zeros((2*N,2))
    Par_off[:,1]=np.arange(2*N)
    
    
    #DECODED VALUE & 'x1,x2...' EVALUATION:: 

    decval_x=np.zeros((N,no_of_var))
    xold=np.zeros((N,no_of_var))
    
    
    for i in range(N):
        for j in range(no_of_var):
            for k in range(0,length[j]):
                decval_x[i][j] = decval_x[i][j] + (2**(length[j]-k-1)) * S[i][k+cumm_len[j]]

    for i in range(N):
        for j in range(no_of_var):
            xold[i][j] = x_min[j] + (x_max[j]-x_min[j])/(2**length[j]-1) * decval_x[i][j]
    
    
        
    #FUNCTION CALLING & FITNESS EVALUATION::     


    funval=np.zeros(N)
    for i in range(N):
        funval[i]= function(xold[i])

    fitness=np.zeros(N)
    fitness = np.array(fitness_eval(funval))
    
    Par_off[0:N,0]=fitness
    
    
    #BINARY TOURNAMENT SELECTION::  
    

    matpool = np.zeros( shape=(N,tot_bit) )
    j=0
    
    #for m in range(2):
     #   indx=np.arange(N)
      #  np.random.shuffle(indx)
        
    #for i in range(0,N,2):
     #   if(fitness[indx[i]]>fitness[indx[i+1]]):
      #      matpool[j]=S[indx[i]]
      #  else:
       #     matpool[j]=S[indx[i+1]]
       # j=j+1
    
    n=2
            
    for i in range(N):
        indx1=np.random.randint(N)
        indx2=np.random.randint(N)
        
    if(fitness[indx1]>fitness[indx2]):
        max_fit = fitness[indx1]
        indx=indx1
        
    else:
        max_fit = fitness[indx2]
        indx=indx2
            
    for j in range(n-2):
        tem=np.random.randint(N)
        if(fitness[tem]>max_fit):
            max_fit = fitness[tem]
            indx=tem
        
    matpool[i]=S[indx]
    
        
    #SINGLE POINT CROSSOVER::      
        
      
    indx=np.arange(N)
    np.random.shuffle(indx)
    
    for i in range(0,N,2):
            p1=np.random.randint(0, tot_bit)
            for j in range(p1,tot_bit):
                matpool[indx[i]][j],matpool[indx[i+1]][j] = matpool[indx[i+1]][j],matpool[indx[i]][j]
          
                    
   #GLOBAL MUTATION::                 
                    
    pm=1/no_of_var
    for i in range(N):
        for j in range(tot_bit):
            probab=np.random.uniform(0,1)
            if(probab<pm):
                if(matpool[i][j]==0):
                    matpool[i][j]=1
                else:
                    matpool[i][j]=0
    
   
    
    decval_x=np.zeros((N,no_of_var))
    xnew=np.zeros((N,no_of_var))
    
    
    for i in range(N):
        for j in range(no_of_var):
            for k in range(0,length[j]):
                decval_x[i][j] = decval_x[i][j] + (2**(length[j]-k-1)) * matpool[i][k+cumm_len[j]]

    for i in range(N):
        for j in range(no_of_var):
            xnew[i][j] = x_min[j] + (x_max[j]-x_min[j])/(2**length[j]-1) * decval_x[i][j]
    
    
        
    #FUNCTION CALLING & FITNESS EVALUATION::     


    funval=np.zeros(N)
    for i in range(N):
        funval[i]= function(xnew[i])

    fitness=np.zeros(N)

    fitness = np.array(fitness_eval(funval))
    
    Par_off[N:2*N,0]=fitness
    Par_off=Par_off[Par_off[:,0].argsort()]
    Par_off=np.flipud(Par_off)
   
    for t in range(N):
        if(Par_off[t][1]<N):
            S_new[t]=S[int(Par_off[t][1])]
        else:
            S_new[t]=matpool[int(Par_off[t][1]-N)]
    
    S=S_new
    if(int(Par_off[0][1])>=N):
        print("generation :",gen," x =",xnew[int(Par_off[0][1])-N])
    else:
        print("generation :",gen," x =",xold[int(Par_off[0][1])])

if(int(Par_off[0][1])>=N):
    ans=xnew[int(Par_off[0][1])-N]
else:
    ans=xold[int(Par_off[0][1])]
print("The value of independent variable where optimum lies: ",ans)
print("optimum function value is: ",function(ans))

    

    
```
